```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()
 
objects = ('Lottery', 'Non-Lottery 1st Round', '2nd Round')
y_pos = np.arange(len(objects))
performance = [33,6,6]
 
plt.bar(y_pos, performance, align='center', alpha=0.5, color=['blue', 'green', 'red'])
plt.xticks(y_pos, objects)
plt.ylabel('All NBA Players')
plt.title('All NBA Players')

plt.show()
```

```python
N = 3
firstTeam = (37, 10, 5)
secondTeam = (30, 10, 11)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, firstTeam, width)
p2 = plt.bar(ind, secondTeam, width,
             bottom=firstTeam)

plt.ylabel('Appearances')
plt.title('All Rookie Team Appearances')
plt.xticks(ind, ('Lottery', 'Non-Lottery 1st Round', '2nd Round'))
plt.yticks(np.arange(0, 80, 10))
plt.legend((p1[0], p2[0]), ('1st Team', '2nd Team'))

plt.show()
```

```python
objects = ('Lottery', 'Non-Lottery 1st Round', '2nd Round')
y_pos = np.arange(len(objects))
percentages = [75.7,42.5,20.6]
 
plt.bar(y_pos, percentages, align='center', alpha=0.5, color=['blue', 'green', 'red'])
plt.xticks(y_pos, objects)
plt.ylabel('Percentage')
plt.title('Percentage of Active NBA Players')

plt.show()
```
